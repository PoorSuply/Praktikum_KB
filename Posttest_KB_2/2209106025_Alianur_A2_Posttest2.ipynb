#####Menampilkan informasi jumlah, nama, dan jumlah attribute bertipe angka didalam Dictionary

```python
#import csv ke dalam python
import csv
# Dictionary untuk menyimpan informasi
attribute_info = {
    'total_attributes': 0,
    'numeric_attributes': 0,
    'attribute_names': [],
    'total_records': 0
}

# Memuat file CSV ke dalam variabel "file_name"
file_name = 'large_data.csv'

with open(file_name, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    # Ambil nama-nama atribut dari header
    attribute_names = csv_reader.fieldnames

    # Hitung jumlah atribut
    total_attributes = len(attribute_names)

    # Inisialisasi hitungan rekaman
    total_records = 0

    # Iterasi melalui baris-baris data
    for row in csv_reader:
        total_records += 1  # Menambahkan 1 ke total rekaman
        for attribute_name, attribute_value in row.items():
            # Cek apakah nilai atribut adalah angka
            if attribute_value.isnumeric():
                attribute_info['numeric_attributes'] += 1

    # Simpan informasi ke dalam dictionary
    attribute_info['total_attributes'] = total_attributes
    attribute_info['attribute_names'] = attribute_names
    attribute_info['total_records'] = total_records  # Menyimpan total rekaman

# Menampilkan informasi
print("Jumlah total atribut:", attribute_info['total_attributes'])
print("Jumlah atribut bertipe angka:", attribute_info['numeric_attributes'])
print("Nama-nama atribut:", attribute_info['attribute_names'])
print("Total record:", attribute_info['total_records'])  # Menampilkan total rekaman
```
