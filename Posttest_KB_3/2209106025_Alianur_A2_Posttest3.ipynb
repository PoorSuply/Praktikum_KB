##### Mengimport Libarary yang akan digunakan 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Memasukan file yang dibaca ke dalam variabel df

```python
df=pd.read_csv('large_data.csv')
```

##### Pie chart/ Diagram lingkaran

```python
sns.set_theme()
dftype = df['TYPE'].value_counts()
plt.pie(dftype,labels=dftype.index,autopct='%1.1f%%',startangle=90,explode=[0.01,0.01,0.01,0.01],shadow=True)
plt.title('Distribusi gejala pada penyakit',fontsize=12,fontweight='bold')
plt.show()
```

 ##### Pie chart ini menunjukan seberapa besar suatu penyakit menyumbangkan atau memunculkan gejala-gejala seperti cough, tiredness, muscle aches, dan lain-lain kepada penderitanya. Dapat disimpulkan bahwa gejala yang muncul 56,2% nya berasal dari penyakit FLU, dan penyakit COLD sebagai penyakit yang paling sedikit memunculkan gejala.

##### Heatmap

```python
data = {'TYPE': np.random.choice(['Covid', 'Flu', 'Cold', 'Allergy'], size=100),
        'Symptom': np.random.choice(['Fever', 'Cough', 'Runny Nose', 'Sneezing'], size=100)}
df = pd.DataFrame(data)
df['TYPE'] = df['TYPE'].astype('category')
contingency_table = pd.crosstab(df['TYPE'], df['Symptom'])
plt.figure(figsize=(12, 8))
sns.heatmap(contingency_table, cmap='viridis', annot=True, fmt='d', linewidths=.5, cbar_kws={'label': 'Count'})
plt.title('Heatmap Korelasi antar TYPE dan Symptom')
plt.show()
```

##### Grafik ini digunakan untuk menunjukan korelasi antara tipe penyakit dengan gejalanya. Semakin kuning warna sel, menunjukan bahwa adanya korelasi positif antara penyakit dengan gejala yang ditimbulkan. semakin ke warna ungu, menunjukan bahwa adanya korelasi negatif antara penyakit dengan gejala yang ditimbulkan. Dari data ini dapat disimpulkan penyakit Covid memiliki kecenderungan sangat tinggi untuk menimbulkan gejala batuk(cough), dan penyakit Alergi memiliki kecenderungan kecil untuk menimbulkan gejala bersin.

##### Grafik Batang

```python
dftitanic = df['Symptom'].value_counts()
plt.figure(figsize=(7,3))
plt.bar(dftitanic.index,dftitanic,color=['red','yellow','brown', 'blue'])
plt.xlabel('Gejala')
plt.ylabel('Jumlah kasus')
plt.title('Jumlah Kasus untuk Setiap Gejala',fontsize=12,fontweight='bold')
plt.show()
```

##### Grafik batang ini menunjukan jumlah kasus gejala cough, fever, runny nose dan sneezing dari dataset. dari data ini dapat disimpulkan bahwa cough/batuk memiliki record total terbanyak dan yang terdikit adalah sneezing.

##### Histogram

```python
data = {'TYPE': np.random.choice(['Covid', 'Flu', 'Cold', 'Allergy'], size=100),
        'Symptom': np.random.choice(['Fever', 'Cough', 'Runny Nose', 'Sneezing'], size=100)}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Symptom', hue='TYPE', multiple='stack', palette='viridis',bins=50)
plt.title('Distribusi Gejala untuk Setiap Tipe')
plt.xlabel('Gejala')
plt.ylabel('Jumlah Kasus')
plt.show()
```

##### Histogram ini menunjukan distribusi gejala pada setiap penyakit, bisa kita lhat penyakit allergy berada diatas untuk setiap gejala. hal itu menunjukan bahwa allergy adalah penyakit yang paling umum untuk menimbulkan berbagai gejala. selain itu, kita bisa lihat dibagian Cough allergy berada diposisi lebih tinggi dibanding dengan kolom gejala lainnya. hal itu menunjukan bahwa Cough adalah gejala yang paling umum muncul pada penyakit allergy.
